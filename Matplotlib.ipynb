```python
import matplotlib.pyplot as plt 
import numpy as np

day = np.arange(1, 16)  #Number of days (1-15)
study_hours = np.random.randint(0, 12, size=15) #Random Study hours for 15 days

diff_colors = plt.cm.rainbow(np.linspace(0, 1, len(day))) #different colors of bars and plot

#legend is function use to display label

#Line Chart
plt.figure(figsize=(8, 5))
plt.title("Line Chart",fontsize=20)
plt.plot(day,study_hours, color='purple', marker="o", label="Hours", alpha=1) #Parameters of Stem function
plt.xlabel("Days", fontsize=15,color="blue")
plt.ylabel("Study Hours", fontsize=15,color="blue")
plt.legend()
plt.grid(True)
plt.show()


#Bar Chart
plt.figure(figsize=(8, 5))
plt.title("Bar Chart",fontsize=20)
plt.bar(day,study_hours, color=diff_colors, label="Hours",width=0.8,align="edge",edgecolor="black",linewidth=1.5) #Parameters of Stem function
plt.xlabel("Days",fontsize=15,color="blue")
plt.ylabel("Study Hours",fontsize=15,color="blue")
plt.legend()
plt.show()


#Scatter Chart
plt.figure(figsize=(8, 5))
plt.title("Scatter Chart", fontsize=20)
plt.scatter(day,study_hours, color=diff_colors, marker='*', s=150 ,label="Hours") #Parameters of Stem function
plt.xlabel("Days", fontsize=15,color="blue")
plt.ylabel("Study Hours", fontsize=15,color="blue")
plt.colorbar()
plt.legend()
plt.grid(True)
plt.show()


#Histogram Chart
plt.figure(figsize=(6, 5))
plt.title("Histogram Chart",fontsize=20)
plt.hist(study_hours, bins="auto",color="orange", label="Hours",edgecolor="purple",align="right") #Parameters of Stem function
plt.xlabel("Study Hours", fontsize=15,color="blue")
plt.ylabel("Frequency", fontsize=15,color="blue")
plt.legend()
plt.show()


#Pie Chart
plt.figure(figsize=(5,5))
plt.title("Pie Chart",fontsize=20,loc="center")
plt.pie(study_hours, labels=day, autopct="%0.1f%%", shadow=True, radius=2, labeldistance=1.1, counterclock=False,pctdistance=0.9) #Parameters of Stem function
plt.legend(loc="best")
plt.show()


#Boxplot Chart
plt.figure(figsize=(8, 5))
plt.title("Boxplot",fontsize=20)
plt.boxplot(study_hours,notch=False,widths=0.2,labels=["Study Hours"],patch_artist=True,showmeans=True) #Parameters of Stem function
plt.show()


#Stem Chart
plt.figure(figsize=(8, 5))
plt.title("Stem Plot", fontsize=20)
plt.stem(day, study_hours,linefmt=":", markerfmt="go",basefmt="b",bottom=5, label="Hours") #Parameters of Stem function
plt.xlabel("Days", fontsize=15,color="blue")
plt.ylabel("Study Hours", fontsize=15,color="blue")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()




```
